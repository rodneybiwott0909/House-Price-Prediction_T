```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv('house_prices.csv')
```

```python
# Dataframe preview
print(df.head())
```

```python
# Dataset shape
print(f"Rows: {df.shape[0]} \nColumns: {df.shape[1]}")
```

```python
# Dataset Information
print(df.info())
```

```python
# Statistical Summary
print(df.describe())
```

```python
# Missing values
print(df.isnull().sum())
```

### Numpy Analysis

```python
# convert to numpy arrays
prices = df['house_price'].values
square_feet = df['square_feet'].values
```

```python
### Numpy Analysis
print(f"Mean Price: Ksh. {np.mean(prices):,.2f}")
print(f"Median Price: Ksh. {np.median(prices):,.2f}")
print(f"Std Deviation: Ksh. {np.std(prices):,.2f}")
print(f"Min Price: Ksh. {np.min(prices):,.2f}")
print(f"Max Price: Ksh. {np.max(prices):,.2f}")
print(f"Price range: Ksh. {np.ptp(prices):,.2f}")
```

### Data Visualization

```python
# Price distribution

plt.figure(figsize = (12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['house_price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('House Price (Ksh.)')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')

plt.subplot(1, 2, 2)
plt.boxplot(df['house_price'])
plt.ylabel('House Price (Ksh.)')
plt.title('House Prices Boxplot')

plt.tight_layout()
plt.savefig('price_distribution.png', dpi=3000, bbox_inches='tight')
plt.show()
```

```python
## Scatter plots for all features

fig, axes = plt.subplots(2, 3, figsize = (15, 10))

#  Square feet vs Price
axes[0, 0].scatter(df['square_feet'], df['house_price'], alpha=0.6, color='blue')
axes[0, 0].set_xlabel('Square Feet')
axes[0, 0].set_ylabel('House Price (Ksh.)')
axes[0, 0].set_title('Square feet vs Price')

#  Bedroom vs Price
axes[0, 1].scatter(df['bedrooms'], df['house_price'], alpha=0.6, color='green')
axes[0, 1].set_xlabel('Bedrooms')
axes[0, 1].set_ylabel('House Price (Ksh.)')
axes[0, 1].set_title('Bedrooms vs Price')

#  Bathrooms vs Price
axes[0, 2].scatter(df['bathrooms'], df['house_price'], alpha=0.6, color='red')
axes[0, 2].set_xlabel('Bathrooms')
axes[0, 2].set_ylabel('House Price (Ksh.)')
axes[0, 2].set_title('Bathrooms vs Price')

#  Age vs Price
axes[1, 0].scatter(df['age_years'], df['house_price'], alpha=0.6, color='orange')
axes[1, 0].set_xlabel('Age (Years)')
axes[1, 0].set_ylabel('House Price (Ksh.)')
axes[1, 0].set_title('Age vs Price')

#  Garage vs Price
axes[1, 1].scatter(df['garage'], df['house_price'], alpha=0.6, color='brown')
axes[1, 1].set_xlabel('Garage Capacity')
axes[1, 1].set_ylabel('House Price (Ksh.)')
axes[1, 1].set_title('Garage vs Price')

#  Garage vs Price
axes[1, 2].scatter(df['location_score'], df['house_price'], alpha=0.6, color='purple')
axes[1, 2].set_xlabel('Location Score')
axes[1, 2].set_ylabel('House Price (Ksh.)')
axes[1, 2].set_title('Location Score vs Price')

plt.tight_layout()
plt.savefig('feature_relationship.png', dpi=3000, bbox_inches='tight')
plt.show()
```
